```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
students_names=["sanjay","Rahul","Karan","Wasim","Ramesh","Ajay","Sartaj","Priya"]
students_marks=[35,50,20,45,25,40,25,40]
```

```python
#Create a dictionary of students names and marks
students_marks_dict = {"students_names":students_names,"students_marks":students_marks}
print(students_marks_dict)
print(type(students_marks_dict))
print(len(students_marks_dict))
print(students_marks_dict.keys())
print(students_marks_dict.values())
print(students_marks_dict.items())
```

```python
# 2. Create a dataframe from this dictionary
students_marks_df = pd.DataFrame(students_marks_dict)
print(students_marks_df)
```

```python
marks_perc = []
for x in students_marks:
	res = (x/50)*100
	marks_perc.append(res)

print(marks_perc)
```

# **Line Chart**

```python
def line_chart_of_students_and_marks():
  plt.plot(students_names,students_marks)
  plt.title("Students Marks Graph")
  plt.xlabel("Students Names")
  plt.ylabel("Students Marks")
  plt.show()

line_chart_of_students_and_marks()
```

# **Bar Chart**

```python
def percentage_bar_chart():
  plt.bar(students_names,marks_perc)
  plt.title("Students' Percentage Graph")
  plt.xlabel("Student Names")
  plt.ylabel("Student Percentage")
  plt.show()

percentage_bar_chart()
```
